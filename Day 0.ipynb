```python
#electricity Bill Calculator
units = int(input("Enter the no. of units: "))
print("Units: " , units)
if units <= 100:
    print("Electricity Bill: ₹" , units * 5)
    print("Surcharge: NO")
elif units <= 200:
    print("Electricity Bill: ₹" , (100 * 5) + (units - 100) * 7)
    print("Surcharge: NO")
elif units <= 300:
    print("Electricity Bill: ₹" , (100 * 5) + (100 * 7) + (units - 200) * 10)
    print("Surcharge: NO")
else:
    print("Electricity Bill: ₹" , (100 * 5) + (100 * 7) + (100 * 10) + ((units - 300) * 12) + 500)
    print("Surcharge: YES")
```

```python
#ATM Withdrawl Validator
a = int(input("Balance: "))
b = int(input("Witndrawl: "))
if a >= b:
    if b % 100 == 0:
        print("Withdrawl Successful")
        print("Amount Withdrawn: ₹" , b)
        print("Remaining Balance: ₹" , a - b - 10)
    else:
        print("Withdrawl Unsuccessful")
        print("Amount must be a multiple of 100.")
else:
    print("Withdrawl Unsuccessful")
    print("Not enough balance.")
```
