```python
a = input("Enter a password: ")
b = len(a)
def has_digit(s):
    return any(ch.isdigit() for ch in s)

def has_upper(s):
    return any(ch.isupper() for ch in s)

c = has_digit(a)
d = has_upper(a)
if b >= 8 and c == True and d == True:
    print("Strong Password")
elif b >= 8 and c == True:
    print("Moderate Password")
elif b >= 8 and d == True:
    print("Moderate Password")
elif c == True and d == True:
    print("Moderate Password")
else:
    print("Weak Password") 
```
